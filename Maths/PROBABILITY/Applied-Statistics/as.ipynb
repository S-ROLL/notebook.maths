# Kiểm định giả thuyết

> - Kiểm định 1 phía (cao hơn hoặc thấp hơn).
> - Kiểm định 2 phía (giống nhau hoặc khác nhau).

- **B1.** Gọi giả thuyết $H_1$ và $H_0$
- **B2.** So sánh giá trị `sig.`
  - `sig.`$>5\% \Rightarrow$  PSĐN $\Rightarrow$ Dùng dòng 1 của bảng 1.2 để so sánh tiếp $\Rightarrow$ Chấp nhận $H_0$
  - `sig.`$<5\% \Rightarrow$  PSKĐN $\Rightarrow$ Dùng dòng 2 của bảng 1.2 để so sánh tiếp $\Rightarrow$ Chấp nhận $H_1$
- **B3.** Tính $p - \text{value}$
  - 1 phía $\Rightarrow p - \text{value} = \Large \frac{\text{sig. (2-tailed)}}{2}$
  - 2 phía $\Rightarrow p - \text{value} = \text{sig. (2-tailed)}$
- **B4.** $p - \text{value} > 5\% \Rightarrow$ chấp nhận $H_0$ bác bỏ $H_1$ và ngược lại.
- **B5.** Đưa ra cặp giả thiết mới (cho bài toán 1 phía)
- **B6.** Đánh giá $ p - \text{value} = \Large \frac{\text{sig. (2-tailed)}}{2}$ mới
- **B7.** KL

# Binomial Test

### Thông tin

- Category : Loại dữ liệu.
- N : Số lượng quan sát được.
- Observed Prop. : tỉ lệ quan sát được.
- Test Prop. : tỉ lệ giả thuyết.
- Exact Sig. (1-tailed) : Mức ý nghĩa (kiểm định 1 phía).

### Nguồn gốc

$$
\begin{split}
N &= N_1 + N_2 \\
\text{Observed Prop.} &= \frac{N_i}{\text{N total}}\\
\end{split}
$$

# One-sample t-test

### Thông tin

<p style="text-align: center;">Bảng 1</p>

- N : số lượng quan sát được hoặc người tham gia.
- Mean : giá trị trung bình.
- Std. Deviation : độ lệch chuẩn.
- Std. Error Mean: độ sai số trung bình.

<p style="text-align: center;">Bảng 2</p>

- t : thống kê sử dụng kiểm định phương sai.
- df : bậc tự do.
- Sig. (2-tailed) : mức ý nghĩa (kiểm định 2 phía)
- Mean Difference : trung bình mẫu.
- 95% Confidence Interval of the Difference : khoảng tin cậy 95% của mẫu.
  - Lower : giới hạn dưới.
  - Upper : giới hạn trên.

### Nguồn gốc

$$
\begin{split}
\text{Std. Error Mean} &= \frac{\text{Std. Deviation}}{\sqrt{N}} \\
df &= N -1 \\
\end{split}
$$

# Independent-samples t-test

### Thông tin

<p style="text-align: center;">Bảng 1.1</p>

- N : số biến quan sát năng suất của 2 giống lúa A và B.
- Mean : giá trị trung bình về năng suất của 2 giống lúa.
- Std. Deviation : độ lệch chuẩn về năng suất của 2 giống lúa. 
- Std. Error Mean : độ sai số trung bình về năng suất của từng giống lúa.

<p style="text-align: center;">Bảng 1.2</p>

- Leneve's Test for Equality of Variances : kiểm định phương sai bằng kiểm định Leneve.
- F : thống kê F.
- Sig : mức ý nghĩa.
- t-test for Equality of Mean : kiểm định trung bình bởi t-test.
- t : thống kê sử dụng để kiểm định phương sai.
- df : bậc tự do.
- sig. (2-tailed) : mức ý nghĩa (kiểm định 2 phía).
- Mean Differences : trung bình mẫu.
- Std. Error Difference : độ sai số của mẫu.
- 95% Confidence Interval of the Difference : khoảng tin cậy 95% của mẫu.
  - Lower : giới hạn dưới.
  - Upper : giới hạn trên.
- Equal variances assumed : phương sai đồng nhất.
- Equal variances not assumed : phương sai không đồng nhất.

### Nguồn gốc

$$
\begin{split}
\text{Std. Error Mean } &= \frac{\text{Std. Deviation}}{\sqrt{N}} \\
s^2 &= \frac{(n-1)s_1^2 + (m-1)s_2^2}{n+m-2} \\
t &= \frac{\overline{x}_1 -\overline{x}_2}{\sqrt{\frac{s^2}{n} + \frac{s^2}{m}}} \\
df &= m + n -2
\end{split}
$$

# Paired-samples t-test

### Thông tin

<p style="text-align: center;">Bảng 2.1</p>

- Mean : giá trị trung bình mức độ hài lòng của phụ huynh.
- N : số biến quan sát mức độ hài lòng.
- Std. Deviation : độ lệch chuẩn về mức độ hài lòng.
- Std Error Mean : độ sai số trung bình về mức độ hài lòng.

<p style="text-align: center;">Bảng 2.2</p>

- 95% Confidence Interval of Difference : khoảng tin cậy 95%.
  - Lower : giới hạn dưới.
  - Upper : giới hạn trên.
- t : thống kê sử dụng kiểm định phương sai.
- df : bậc tự do.
- Sig (2-tailed) : mức ý nghĩa (kiểm định 2 phía).

### Nguồn gốc

$$
\begin{split}
\text{std. Error Mean } &= \frac{\text{std. Deviation}}{\sqrt{N}} \\
\text{Mean } &= \overline{x}_1 - \overline{x}_2 \\
t &= \frac{\text{Mean}\sqrt{N}}{\text{std. Deviation}} \\
df &= N -1 \\
\end{split}
$$
